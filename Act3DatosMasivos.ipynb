#### Modelos de análisis de datos
1. Supervisados
    
    a. Clasificación:
        • Árboles de decisión
          * ¿Cómo funcionan? Los árboles de decisión son una herramienta de aprendizaje automático que se utiliza para resolver problemas en los que se deben clasificar o predecir cosas.
          * ¿Cuándo usarlos? Son útiles cuando quieres entender cómo las características de entrada se relacionan con las salidas, funcionan especialmente bien cuando estas relaciones son complicadas y no siguen una línea recta.
        • Regresión logística
          * ¿Cómo funciona? La regresión logística se usa cuando deseas clasificar cosas en una de dos categorías o más, calcula la probabilidad de que una observación pertenezca a una categoría específica, después usa una fórmula matemática llamada función logística para convertir la información de entrada en una probabilidad entre 0 y 1.
          * ¿Cuándo usarla? Se utiliza cuando necesitas predecir una variable que tiene categorías y es muy común en campos como el marketing, la medicina y la biología.
    b. Regresión:
        • Regresión lineal
          * ¿Cómo funciona? La regresión lineal se usa para predecir un número basado en uno o más números de entrada, es decir que busca una relación matemática simple entre las entradas y la salida.
          * ¿Cuándo usarla? Se utiliza cuando deseas entender y predecir relaciones sencillas entre variables; es muy común en análisis de datos y estadísticas.

    c. Series de tiempo:
        • ARIMA
          * ¿Cómo funciona? ARIMA es un modelo estadístico que se utiliza para analizar y predecir datos que cambian con el tiempo, como las series de tiempo, combina diferentes patrones estadísticos para capturar tendencias y patrones temporales.
          * ¿Cuándo usarlo? Se utiliza para analizar y predecir series de tiempo, como las finanzas o el clima.
2. No supervisados
    a. Agrupamiento o clustering:
        • K-means
          * ¿Cómo funciona? K-Means es un algoritmo que agrupa datos en grupos basados en similitudes. Intenta encontrar grupos de datos que sean similares entre sí y diferentes de otros grupos, lo hace midiendo la distancia entre los puntos de datos.
          * ¿Cuándo usarlo? Se usa cuando quieres agrupar datos en categorías basadas en similitudes, por ejemplo, en segmentación de clientes o análisis de redes sociales.
    b. Reglas de asociación:
        • A priori
          * ¿Cómo funciona? El algoritmo A priori busca patrones en conjuntos de datos, identificando elementos que suelen aparecer juntos en transacciones o registros. Luego, evalúa cuán importantes son estos patrones.
          * ¿Cuándo usarlo? Se usa en aplicaciones como recomendación de productos y análisis de patrones de compra en comercio electrónico.
